# Import thư viện cần thiết 

```python
import pandas as pd
```

### 1.Import file CSV


```python
data = pd.read_csv("data/Telco-Customer-Churn.csv")
data
```

```python
data.info()
```

### 2. Tiền xử lý dữ liệu 

```python
# Xử lý giá trị missing 
# kiểm tra tổng giá trị missing 
data_null = data.isnull().sum()
print(data_null)
```

#### Sau kiểm tra thì không có giá trị null nào. Nếu có thì hãy chạy cell bên dưới

```python
data_no_null = data.dropna(inplace=False)
data_no_null.info()
```

```python
# Loại bỏ duplicates dựa trên customerID

data_duplicates = data_no_null.copy()

data_duplicates["customerID"].value_counts().sort_index(ascending = False) # Đếm số lượng values

data_duplicates_customerID = data_duplicates.drop_duplicates("customerID") # drop duplicates
data_duplicates_customerID.info()
```

```python
# Chuyển đổi SentiorCitizen 
data_sentiorcitizen = data_no_null.copy() # copy data_no_null

data_sentiorcitizen["SeniorCitizen"].value_counts().sort_index() # Kiểm tra giá trị của SeniorCitizen

# Chuyển đổi sang Yes/NO

data_sentiorcitizen["SeniorCitizen"] = data_sentiorcitizen["SeniorCitizen"].map({0:"No", 1:"Yes"}) #map 0 --> No, 1 --> Yes
data_sentiorcitizen["SeniorCitizen"].value_counts().sort_index() # Kiểm tra giá trị sau khi chuyển đỗi

data_sentiorcitizen.head()
```

```python
# Tạo cột mới Tenure Group (nhóm thời gian: <12 tháng, 12-36, >36)
data_tenure = data_sentiorcitizen.copy()

# Cột InternetServices chứa thời gian sử dụng internet nên sẽ nhóm theo đó 
data_tenure["tenure"].value_counts().sort_index() # Kiểm tra giá trị của tenure

def tenure_group(tenure):
    if tenure < 12:
        return "<12 months"
    elif tenure >= 12 and tenure <= 36:
        return "12-36 months"
    else:
        return ">36 months"

data_tenure["tenure_group"] = data_tenure["tenure"].apply(tenure_group) # Áp dụng hàm tenure_group vào cột tenure để tạo cột tenure_group
data_tenure["tenure_group"].value_counts().sort_index() # Kiểm tra giá trị của tenure_group
```

```python
data_tenure.head()
```

```python
# Tạo cột mới Total Services (đếm số dịch vụ như PhoneService, MultipleLines, InternetService)
data_services = data_tenure.copy()
# Cột dịch vụ sẽ là các cột có giá trị Yes/No, nên sẽ đếm số cột có giá trị Yes
service_columns = ["PhoneService", "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"] # Các cột dịch vụ
data_services["TotalServices"] = data_services[service_columns].apply(lambda x: x == "Yes").sum(axis=1) # Đếm số cột có giá trị Yes và tạo cột TotalServices
data_services["TotalServices"].value_counts().sort_index() # Kiểm tra giá trị của TotalServices
```

```python
data_services.head()
```

```python
# Lưu data đã được làm sạch vào file mới
data_services.to_csv("data/Telco-Customer-Churn-Cleaned.csv", index=False)
```
